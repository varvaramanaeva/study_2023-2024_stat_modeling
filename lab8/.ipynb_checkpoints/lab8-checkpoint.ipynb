## Линейное программирование

```julia
using JuMP
using GLPK
```

```julia
# Определение объекта модели с именем model:
model = Model(GLPK.Optimizer)
```

```julia
# Определение переменных x, y и граничных условий для них:
@variable(model, x >= 0)
@variable(model, y >= 0)
```

```julia
# Определение ограничений модели:
@constraint(model, 6x + 8y >= 100)
@constraint(model, 7x + 12y >= 120)
```

```julia
# Определение целевой функции:
@objective(model, Min, 12x + 20y)
```

```julia
# Вызов функции оптимизации:
optimize!(model)
```

```julia
# Определение причины завершения работы оптимизатора:
termination_status(model)
```

```julia
# Демонстрация первичных результирующих значений переменных x и y:
@show value(x)
@show value(y)
# Демонстрация результата оптимизации:
@show objective_value(model)
```

## Оптимизация рациона

```julia
category_data = JuMP.Containers.DenseAxisArray(
    [1800 2200;
    91 Inf;
    0 65;
    0 1779],
    ["calories", "protein", "fat", "sodium"],
    ["min", "max"])
```

```julia
# массив данных с наименованиями продуктов:
foods = ["hamburger", "chicken", "hot dog", "fries", "macaroni", "pizza","salad", "milk", "ice cream"]
```

```julia
# Массив стоимости продуктов:
cost = JuMP.Containers.DenseAxisArray(
[2.49, 2.89, 1.50, 1.89, 2.09, 1.99, 2.49, 0.89, 1.59],
foods)
```

```julia
food_data = JuMP.Containers.DenseAxisArray(
    [410 24 26 730;
    420 32 10 1190;
    560 20 32 1800;
    380 4 19 270;
    320 12 10 930;
    320 15 12 820;
    320 31 12 1230;
    100 8 2.5 125;
    330 8 10 180],
    foods,
    ["calories", "protein", "fat", "sodium"])
```

```julia
# Определение объекта модели с именем model:
model_calories = Model(GLPK.Optimizer)
```

```julia
# Определим массив:
categories = ["calories", "protein", "fat", "sodium"]
```

```julia
# Определение переменных:
@variables(model_calories, begin
    category_data[c, "min"] <= nutrition[c = categories] <= category_data[c, "max"]
    # Сколько покупать продуктов:
    buy[foods] >= 0
    end)
```

```julia
# Определение целевой функции:
@objective(model_calories, Min, sum(cost[f] * buy[f] for f in foods))
```

```julia
# Определение ограничений модели:
@constraint(model_calories, [c in categories],
sum(food_data[f, c] * buy[f] for f in foods) == nutrition[c])
```

```julia
# Вызов функции оптимизации:
JuMP.optimize!(model_calories)
term_status = JuMP.termination_status(model_calories)
```

```julia
hcat(buy.data,JuMP.value.(buy.data))
```

## Путешествие по миру

```julia
using DelimitedFiles
using CSV
```

```julia
passportdata = readdlm("data/passport-index-matrix.csv",',')
```

```julia
# Задаём переменные:
cntr = passportdata[2:end,1]
vf = (x -> typeof(x)==Int64 || x == "VF" || x == "VOA" ? 1 : 0).(passportdata[2:end,2:end])
```

```julia
# Определение объекта модели с именем model:
model_passports = Model(GLPK.Optimizer)
```

```julia
# Переменные, ограничения и целевая функция:
@variable(model_passports, pass[1:length(cntr)], Bin)
@constraint(model_passports, [j=1:length(cntr)], sum( vf[i,j]*pass[i] for i in 1:length(cntr)) >= 1)
@objective(model_passports, Min, sum(pass))
```

```julia
# Вызов функции оптимизации:
JuMP.optimize!(model_passports)
termination_status(model_passports)
```

```julia
print(JuMP.objective_value(model_passports)," passports: ",join(cntr[findall(JuMP.value.(pass) .== 1)],", "))
```

## Портфельные инвестиции

```julia
using DataFrames
using XLSX
using Plots
pyplot()
using Convex
using SCS
using Statistics
```

```julia
# Считываем данные и размещаем их во фрейм:
T = DataFrame(XLSX.readtable("data/stock_prices.xlsx","Sheet2"))
```

```julia
# Построение графика:
plot(T[!,:MSFT],label="Microsoft")
plot!(T[!,:AAPL],label="Apple")
plot!(T[!,:FB],label="FB")
```

```julia
# Данные о ценах на акции размещаем в матрице:
prices_matrix = Matrix(T)
```

```julia
# Вычисление матрицы доходности за период времени:
M1 = prices_matrix[1:end-1,:]
M2 = prices_matrix[2:end,:]
# Матрица доходности:
R = (M2.-M1)./M1
```

```julia
# Матрица рисков:
risk_matrix = cov(R)
# Проверка положительной определённости матрицы рисков:
isposdef(risk_matrix)
```

```julia
# Доход от каждой из компаний:
r = mean(R,dims=1)[:]
```

```julia
# Вектор инвестиций:
x = Variable(length(r))
```

```julia
# Объект модели:
problem = minimize(Convex.quadform(x,risk_matrix),[sum(x)==1;r'*x>=0.02;x.>=0])
```

```julia
# Находим решение:
solve!(problem, SCS.Optimizer)
```

## Восстановление изображения

```julia
using Images
```

```julia
# Считывание исходного изображения:
Kref = load("data/khiam-small.jpg")
```

```julia
K = copy(Kref)
p = prod(size(K))
missingids = rand(1:p,400)
```

```julia
K[missingids] .= RGBX{N0f8}(0.0,0.0,0.0)
K
Gray.(K)
```

```julia
# Матрица цветов:
Y = Float64.(Gray.(K))
```

```julia
correctids = findall(Y[:].!=0)
X = Convex.Variable(size(Y))
problem = minimize(nuclearnorm(X))
problem.constraints += X[correctids]==Y[correctids]
```

```julia
# Находим решение:
solve!(problem, SCS.Optimizer)
```

```julia
@show norm(float.(Gray.(Kref))-X.value)
@show norm(-X.value)
colorview(Gray, X.value)
```

## Оптимальная рассадка по залам

```julia
using Random
```

```julia
zals_str = collect(1:5)
zals_data = JuMP.Containers.DenseAxisArray(
    [180 250;
    180 250;
    220 220;
    180 250;
    180 250],
    zals_str,
    ["min", "max"])
```

```julia
# Переделаны обозначения, потому что не нашла способа для оптимизации
N = 1000
peopl = collect(1:N)
people_pref = copy(hcat([shuffle([1, 2, 3, 10000, 10000]) for i in peopl]...))
```

```julia
model_zal = Model(GLPK.Optimizer)
```

```julia
@variable(model_zal, answ[peopl, zals_str], Bin)
```

```julia
for i in peopl
    @constraint(model_zal, sum(answ[i, :]) == 1)
end
for i in zals_str
    @constraint(model_zal, zals_data[i, "min"] <= sum(answ[:, i]) <= zals_data[i, "max"])
end
```

```julia
@objective(model_zal, Min, sum([sum([answ[t, c]*people_pref[c, t] for c in zals_str]) for t in peopl]))
```

```julia
optimize!(model_zal)
```

```julia
termination_status(model_zal)
```

```julia
ress = value.(answ)
```

```julia
zals_filling = zeros(5)
recomendationss = zeros(N)
for i in peopl
    for j in zals_str
        zals_filling[j] += ress[i, j]
        if ress[i, j] == 1
            recomendationss[i] = j
        end
    end
end
```

## План приготовления кофе

```julia
model = Model(GLPK.Optimizer)
@variable(model, raf >= 0)
@variable(model, cappuccino >= 0)
```

```julia
const grain_limit = 500
@constraint(model, raf * 40 + cappuccino * 30 <= grain_limit)
```

```julia
const milk_limit = 2000
@constraint(model, raf * 140 + cappuccino * 120 <= milk_limit)
```

```julia
const sugar_limit = 40
@constraint(model, raf * 5 == sugar_limit)
```

```julia
objective = 400 * raf + 300 * cappuccino
@objective(model, Max, objective)
```

```julia
optimize!(model)
```

```julia
println("Раф кофе: ", round(value(raf)))
println("Капучино: ", round(value(cappuccino)))
println("Прибыль: ", value(objective))
```
